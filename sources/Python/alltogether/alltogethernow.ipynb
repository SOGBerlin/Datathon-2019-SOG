```python
def cleanColNames(df):
    colnames = list(df.columns.values)
    for colname in colnames:
        colnamestripped=colname.strip()
        if colnamestripped.startswith("Jahr"):
            df.rename(index=str, columns={colname: "Jahr"}, inplace=True)
        elif colnamestripped.startswith("Kommentar"):
            df.rename(index=str, columns={colname: "Kommentar"}, inplace=True)
        elif len(colnamestripped) != len(colname):
            df.rename(index=str, columns={colname: colnamestripped}, inplace=True)
    return df
```

```python
def prepFlatten(df, year, colNameDict):
    regex_frage =  re.compile('Frage \d+')
    numrows,_ = df.shape
    dfcopy = df.copy(deep=True)
    for colname in list(dfcopy.columns.values):
        if colname.startswith('Personen-ID'):
            dfcopy.rename(index=str, columns={colname: "!!Personen-ID"}, inplace=True)
        elif colname.startswith('Frage'):
            match=regex_frage.match(colname)
            if match:
                frage= match.group(0)
                mappedFrage = colNameDict[frage]
                dfcopy.rename(index=str, columns={colname: mappedFrage+"_"+year}, inplace=True)
            else:
                print("Error: Found unmatched ", colname, " in ", year) 
        elif colname.startswith("Kommentar"):
            dfcopy.rename(index=str, columns={colname: "Kommentar_"+year}, inplace=True)
        elif colname.startswith("Jahr"):
            dfcopy.drop(columns=[colname], inplace=True)
        else:
            dfcopy.drop(columns=[colname], inplace=True)
            
    dfcopy["data_available_"+year]=True
    return dfcopy
```
